```python
import pandas as pd

# Pigmento
# Ruta al archivo de Excel (asegúrate de ajustar la ruta a tu archivo)
file_path = "C:\\Users\\ArnaldoCarrasco\\Downloads\\202505.xlsx"
# Cargar el archivo de Excel en un DataFrame
cantidad = pd.read_excel(file_path, sheet_name="Cantidad")
print("Archivo cargado con éxito.")
importe = pd.read_excel(file_path, sheet_name="Importe")
print("Archivo cargado con éxito.")
# todo = pd.read_excel(file_path, sheet_name="Ventas por Fabricante")
# print("Archivo cargado con éxito.")
```

```python
cant_melted = cantidad.melt(id_vars=['Fabricante', 'Producto','ean','id'],
                    var_name='Punto de Venta',
                    value_name='Cantidad')
cant_melted['Cantidad'] = cant_melted['Cantidad'].fillna(0).astype(int)
cant_melted['Cantidad'] = cant_melted['Cantidad'].astype(int)
cant_melted['ean'] = cant_melted['ean'].astype(str)

cant_melted
```

```python
importe_melted = importe.melt(id_vars=['Fabricante', 'Producto','ean','id'],
                            var_name='Punto de Venta',
                            value_name='Importe Neto')
importe_melted['Importe Neto'] = importe_melted['Importe Neto'].fillna(0).astype(int)
importe_melted['Importe Neto'] = importe_melted['Importe Neto'].astype(int)
importe_melted['ean'] = importe_melted['ean'].fillna(0).astype(int)
importe_melted['ean'] = importe_melted['ean'].astype(int)

importe.info()
```

```python
df_combinado = pd.merge(cant_melted, importe_melted, on="id", how="inner")
df_combinado

```
