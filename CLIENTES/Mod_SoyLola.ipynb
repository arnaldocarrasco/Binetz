```python
import pandas as pd

# Pigmento
# Ruta al archivo de Excel (asegúrate de ajustar la ruta a tu archivo)
file_path = r"C:\Users\ArnaldoCarrasco\Downloads\soylola202505.xlsx"


# Cargar el archivo de Excel en un DataFrame
lola = pd.read_excel(file_path)
```

```python
lola
```

```python
lola_melted = lola.melt(id_vars=['Articulo Descripcion','Articulo EAN Uni.','COSTO'],
                    var_name='Punto de Venta',
                    value_name='Cantidad')
lola_melted['Cantidad'] = lola_melted['Cantidad'].fillna(0).astype(int)
lola_melted['Cantidad'] = lola_melted['Cantidad'].astype(int)
lola_melted = lola_melted.drop(lola_melted[lola_melted['Cantidad'] == 0].index)
# Supongamos que tienes un DataFrame llamado "df" con las columnas "Cantidad" y "Costo"
lola_melted["Importe Neto"] = lola_melted["Cantidad"] * lola_melted["COSTO"]
lola_melted = lola_melted.drop("COSTO", axis=1)
lola_melted
```
