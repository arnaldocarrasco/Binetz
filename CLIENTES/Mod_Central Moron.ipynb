```python
import pandas as pd

# Pigmento
# Ruta al archivo de Excel (asegúrate de ajustar la ruta a tu archivo)
file_path = "C:\\Users\\ArnaldoCarrasco\\Downloads\\centraloeste202501.xlsx"

# Cargar el archivo de Excel en un DataFrame
moron = pd.read_excel(file_path)
moron
```

```python
moron_melted = moron.melt(id_vars=['Producto BARRAS', 'Producto DESC','Marca DESC'],
                        var_name='Punto de Venta',
                        value_name='Cantidad')
moron_melted['Cantidad'] = moron_melted['Cantidad'].fillna(0).astype(int)
moron_melted['Cantidad'] = moron_melted['Cantidad'].astype(int)
moron_melted = moron_melted.drop(moron_melted[moron_melted['Cantidad'] == 0].index)
moron_melted
```
