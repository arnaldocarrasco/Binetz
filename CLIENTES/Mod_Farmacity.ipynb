```python
import pandas as pd
import numpy as np
```

```python

# Pigmento
# Ruta al archivo de Excel (aseg√∫rate de ajustar la ruta a tu archivo)
file_path = "C:\\Users\\ArnaldoCarrasco\\Downloads\\FARMA SO.xlsx"

# Cargar el archivo de Excel en un DataFrame
farmacity = pd.read_excel(file_path,sheet_name='BASE')
```

```python
df=farmacity.copy()
df.columns = df.iloc[0]
df = df[1:]
```

```python
columns_to_keep = ['Mes', 'Sucursal COD', 'Marca desc', 'Articulo DESC',
                   'CODIGO DE BARRAS', 'OFFLINE, CANTIDAD', 'OFFLINE, IMPORTE',
                   'ONLINE, CANTIDAD', 'ONLINE, IMPORTE']

df = df[columns_to_keep]
```

```python
df_melted = df.melt(id_vars=['Mes','Marca desc', 'Sucursal COD','Articulo DESC','CODIGO DE BARRAS'],
                          var_name='Parametro',
                          value_name='Valor')
```

```python
# Split the "Parametro" column into two rows
split_df = df_melted['Parametro'].str.split(',', expand=True)
```

```python
# Concatenate the split columns back into the original DataFrame
df_melted = pd.concat([df_melted.drop('Parametro', axis=1), split_df], axis=1)
```

```python
df_melted = df_melted.rename(columns={0: 'ON/OFF', 1: 'Parametro'})
```

```python
pivoted_df = df_melted.pivot_table(
    index=['Mes', 'Sucursal COD', 'Marca desc', 'Articulo DESC', 'CODIGO DE BARRAS', 'ON/OFF'],
    columns='Parametro',
    values='Valor',
    aggfunc='sum'   # o 'sum', 'mean', etc.
).reset_index()
# pivoted_df = pivoted_df[pivoted_df["CANTIDAD"] != 0]
pivoted_df
```

```python
pivoted_df = pivoted_df[pivoted_df[" CANTIDAD"] != 0]
pivoted_df[' CANTIDAD'] = pivoted_df[' CANTIDAD'].astype(int)
pivoted_df[' IMPORTE'] = pivoted_df[' IMPORTE'].astype(int)
pivoted_df['CODIGO DE BARRAS'] = pivoted_df['CODIGO DE BARRAS'].astype(str)

pivoted_df
```

```python
pivoted_df = pivoted_df.rename(columns={'Mes': 'Fecha', 'Sucursal COD': 'Punto de Venta','Marca desc': 'Marca','Articulo DESC':'Desc Prod','CODIGO DE BARRAS':'Ean',' CANTIDAD':'Cantidad',' IMPORTE':'Importe Neto','ON/OFF':'Tipo de venta'})
pivoted_df
```

```python
import pandas as pd

# --- 1) Asegurar que 'fecha' sea datetime ---
# Si tus fechas est√°n en formato d√≠a/mes/a√±o (com√∫n en AR), usa dayfirst=True.
pivoted_df["Fecha"] = pd.to_datetime(pivoted_df["Fecha"], errors="coerce", dayfirst=True)

# Opcional: filtrar al rango especificado (dic-2024 a jun-2025)
mask = (pivoted_df["Fecha"] >= "2024-12-01") & (pivoted_df["Fecha"] < "2025-07-01")
df = pivoted_df.loc[mask].copy()

# --- 2) Agrupar por mes y generar dataframes nombrados 42_YYYYMM ---
dfs_por_mes = {}
for per, grp in df.groupby(df["Fecha"].dt.to_period("M")):
    nombre = f"42_{per.strftime('%Y%m')}"   # ejemplo: 42_202504
    dfs_por_mes[nombre] = grp.reset_index(drop=True).copy()

# (OPCIONAL) crear variables en el entorno global con esos nombres
# Nota: mejor usar el diccionario dfs_por_mes; las variables din√°micas son menos mantenibles.
for nombre, dfi in dfs_por_mes.items():
    globals()[nombre] = dfi

# --- 3) Ejemplos de uso ---
# Acceder a abril 2025 desde el diccionario:
df_abril_2025 = dfs_por_mes["42_202504"]

# O bien, si ejecutaste la parte opcional de globals(), ya existe una variable llamada 42_202504:
# 42_202504.head()

# Ver todos los nombres creados:
print(sorted(dfs_por_mes.keys()))
```
