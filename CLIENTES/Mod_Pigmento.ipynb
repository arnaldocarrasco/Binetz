```python
import pandas as pd

# Rutas a los archivos de Excel
file_paths = {
    'p202504': "C:\\Users\\ArnaldoCarrasco\\Downloads\\p202505.xlsx",
    'ffmm202504':     "C:\\Users\\ArnaldoCarrasco\\Downloads\\ffmm202505.xlsx",
    'cb202504':       "C:\\Users\\ArnaldoCarrasco\\Downloads\\cb202505.xlsx"
}

# Función que procesa un DataFrame: melt, conversión de datos, eliminación de 0 y cálculo de Importe_Neto.
def procesar_df(df):
    # Realiza el melt: se utiliza las columnas 'CODIGO', 'DESCRIP' e 'importe unitario' como id_vars.
    df_melted = df.melt(
        id_vars=['Codigo', 'Descrip', 'importe unitario'],
        var_name='Punto de Venta',
        value_name='Cantidad'
    )
    # Rellenar NaN, convertir a entero y eliminar filas con Cantidad 0
    df_melted['Cantidad'] = df_melted['Cantidad'].fillna(0).astype(int)
    df_melted = df_melted[df_melted['Cantidad'] != 0]

    # Calcular columna Importe_Neto
    df_melted = df_melted.assign(Importe_Neto = df_melted['importe unitario'] * df_melted['Cantidad'])
    return df_melted

# Diccionario para almacenar los DataFrame después del melt
dfs_melted = {}

# Procesamiento de cada archivo
for key, path in file_paths.items():
    df = pd.read_excel(path)
    print(f"Archivo '{key}' cargado con éxito.")
    dfs_melted[key] = procesar_df(df)

# Renombrar DataFrames según convención: nombre variable + "_melted"
prestige_melted = dfs_melted['p202504']
ffmm_melted     = dfs_melted['ffmm202504']
cb_melted       = dfs_melted['cb202504']

# Concatenar los tres DataFrames en uno final
pigmento_enero = pd.concat([prestige_melted, ffmm_melted, cb_melted], ignore_index=True)

# Mostrar el DataFrame final
pigmento_enero["Fecha"]="01/05/2025"
```
